```python
import pandas as pd 
import numpy as np 
# import csv file 
my_df = pd.read_csv("top_insta_influencers_data.csv")
my_df
```

```python
# Count Distinct Values - Descending
```

```python
my_df['country'].value_counts()
```

```python
# Count Distinct Values - Ascending
my_df['country'].value_counts(ascending = True)
```

```python
# Influencer's Country - without NaN
my_df['country'].value_counts()
```

```python
# Influencer's Country - with NaN
my_df['country'].value_counts(dropna=False)
```

```python
# Get Relative Frequency - Percentage
my_df['country'].value_counts(normalize=True)
```

```python
# Get specific Item count 
my_df['country'].value_counts()['United Kingdom']
```

```python
# Count Unique Values - Size
my_df.groupby('country').size()
```

```python
# Count Unique Values - Count 
my_df.groupby('country').count()
```

```python
# get a count of all columns across all columns 
numeric_cols = my_df.select_dtypes(include=['number']).apply(pd.Series.value_counts).fillna(0)
string_cols = my_df.select_dtypes(include=['object']).apply(lambda x: pd.Series(x).value_counts()).fillna(0)
# Combine the results 
combined_counts = pd.concat([numeric_cols, string_cols], axis=1).fillna(0)
combined_counts
```
